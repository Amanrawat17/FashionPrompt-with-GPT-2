## Data Preprocess :

```python
description=[]
pose=[]
image=[]
categories=[]
sub_categories=[]
gender=[]
dataset = h5py.File(filename, "r")

for i in range(dataset['input_pose'].shape[0]):
    if (i%10000 ==0):
        print(i)
    if(dataset['input_pose'][i][0] not in [b'id_gridfs_3',b'id_gridfs_5',b'id_gridfs_6',b'id_gridfs_2']):
        gender.append(dataset['input_gender'][i][0])
        categories.append(dataset['input_category'][i][0])
        description.append(dataset['input_description'][i][0])
        pose.append(dataset['input_pose'][i][0])
        sub_categories.append(dataset['input_subcategory'][i][0])
        image.append(dataset['input_image'][i])
print("Last index processed:", i)
```

```python
description=pd.Series(description)
pose=pd.Series(pose)
sub_categories=pd.Series(sub_categories)
categories=pd.Series(categories)
gender=pd.Series(gender)
```

```python
df=pd.DataFrame()
df['pose']=pose
df['description']=description
df['gender']=gender
df['sub_category']=sub_categories
df['category']=categories
df['image']=image
df
```

```python
import torch
from transformers import GPT2LMHeadModel, Trainer, TrainingArguments, GPT2Tokenizer
from sklearn.model_selection import train_test_split

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)  # Load the tokenizer
model = GPT2LMHeadModel.from_pretrained(model_name, device_map="auto")

# Add a pad token and resize the token embeddings
tokenizer.pad_token = "[PAD]"
model.resize_token_embeddings(len(tokenizer))

# Prepare your dataset (assuming df['final_description'] is your data)
# Split the data into training and validation sets
train_data, val_data = train_test_split(df['final_description'], test_size=0.2, random_state=42)

# Tokenize data
train_encodings = tokenizer(train_data.tolist(), padding=True, truncation=True, return_tensors='pt')
val_encodings = tokenizer(val_data.tolist(), padding=True, truncation=True, return_tensors='pt')

# Define a PyTorch Dataset with labels
class TextDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings

    def __len__(self):
        return len(self.encodings['input_ids'])

    def __getitem__(self, idx):
        item = {key: tensor[idx] for key, tensor in self.encodings.items()}
        item['labels'] = item['input_ids'].clone()  # Use input_ids as labels for loss calculation
        return item

# Create datasets
train_dataset = TextDataset(train_encodings)
val_dataset = TextDataset(val_encodings)

# Training arguments
training_args = TrainingArguments(
    output_dir="./kaggle/working/results",  # Save outputs in the Kaggle working directory
    per_device_train_batch_size=2,          # Adjust based on GPU memory
    per_device_eval_batch_size=2,           # Adjust based on GPU memory
    num_train_epochs=3,
    logging_dir="./kaggle/working/logs",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=2,                     # Keep only the last 2 model checkpoints
    load_best_model_at_end=True,            # Automatically load the best model at the end of training
    fp16=True,                              # Enable mixed-precision training for faster performance on GPUs
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# Start training
trainer.train()

# Save the fine-tuned model and tokenizer in the Kaggle working directory
model.save_pretrained("./kaggle/working/fine_tuned_gpt2")
tokenizer.save_pretrained("./kaggle/working/fine_tuned_gpt2")

print("Fine-tuning completed and model saved.")
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the fine-tuned model and tokenizer
model_path = "/kaggle/working/kaggle/working/fine_tuned_gpt2"
model = GPT2LMHeadModel.from_pretrained(model_path)
tokenizer = GPT2Tokenizer.from_pretrained(model_path)

# Define a function to generate text
def generate_text(prompt, max_length=50, num_return_sequences=1):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        num_return_sequences=num_return_sequences,
        do_sample=True,      # Enable sampling to introduce randomness
        top_k=50,            # Only consider the top 50 tokens by probability
        top_p=0.95,          # Only consider tokens with cumulative probability above 95%
        temperature=0.7,     # Control randomness; lower means less random
        pad_token_id=tokenizer.eos_token_id  # Avoid warnings for unpadded sequences
    )
    return [tokenizer.decode(output, skip_special_tokens=True) for output in outputs]

# Sample prompts for testing
sample_prompts = [
    "I'm looking for men's jeans that are"
]

# Generate and print output for each sample prompt
for prompt in sample_prompts:
    print(f"\nPrompt: {prompt}")
    generated_text = generate_text(prompt, max_length=200, num_return_sequences=1)
    print("Generated Text:", generated_text[0])
```
